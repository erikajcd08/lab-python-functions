```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
```

```python
def initialize_inventory(products_list):
    inventory_dict = {}
    print("--- Setup Inventory ---")
    for item in products_list:
        quantity = int(input(f"Enter the quantity for {item}, please: "))
        inventory_dict[item] = quantity
    return inventory_dict
```

```python
def get_customer_orders():
    customer_orders = set()
    another_product = "yes"
    print("\n--- Place Your Orders ---")
    while another_product.lower() in ["yes","y"]:
        choice = input("Enter the name of the product to order: ").strip().lower()
        customer_orders.add(choice)
        another_product = input("Do you want to add another product? (yes/no): ")
    return customer_orders
```

```python
def calculate_order_statistics(customer_orders, products):
    total_orders = len(customer_orders)
    percentage = (total_orders / len(products)) * 100
    return total_orders, percentage
```

```python
def print_order_statistics(order_statistics):
    total, percentage = order_statistics
    print("\n--- Order Summary ---")
    print(f"Total Unique Products Ordered: {total}")
    print(f"Percentage of  Unique Products Ordered: {percentage:.2f}%")
```

```python
def update_inventory(customer_orders, inventory):
    print("\n--- Processing Inventory Updates ---")
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1
            print(f"Stock updated for: {product}")
        else:
            print(f"Note: {product} was not in the catalog.")
    return inventory
```

```python
def print_updated_inventory(inventory):
    print("\n--- Final Updated Inventory ---")
    for product, quantity in inventory.items():
        print(f"Product: {product:12} | Stock Remaining: {quantity}")
```

```python
def main():
    current_inventory = initialize_inventory(products)
    orders = get_customer_orders()
    stats = calculate_order_statistics(orders, products)
    print_order_statistics(stats)
    current_inventory = update_inventory(orders, current_inventory)
    print_updated_inventory(current_inventory)
```

```python
main()
```
